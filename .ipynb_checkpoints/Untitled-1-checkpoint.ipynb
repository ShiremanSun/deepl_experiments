第一题：

```python
import torch
M = torch.tensor([[1.0, 2.0, 3.0]])  # 1x3 矩阵
N = torch.tensor([[1.0], [2.0]])     # 2x1 矩阵


# 方式一：直接减，进行了自动广播
try:
    result1 = M - N
    print("形式1的结果:\n", result1)
except Exception as e:
    print("形式1的报错:\n", e)
    
# 方式二：进行手动广播
try:
    M_expanded = M.expand(2, 3)
    N_expanded = N.expand(2, 3)
    result2 = M_expanded - N_expanded
    print("形式2的结果:\n", result2)
except Exception as e:
    print("形式2的报错:\n", e)
    
#方式三：没有广播的减法
try:
    result3 = M.squeeze() - N.squeeze()
    print("形式3的结果:\n", result3)
except Exception as e:
    print("形式3的报错:\n", e)
```

第二题：

```python
import torch
P = torch.normal(0, 0.01, (3, 2))
Q = torch.normal(0, 0.01, (4, 2))
print(P)
# 求转置
Q_T = Q.t()
print(Q_T)
# 求乘积
result = torch.matmul(P, Q_T)
print(result)
```

```python
# 梯度计算
import torch
x = torch.tensor(1.0, requires_grad=True)
y1 = x**2
with torch.no_grad():
    y2 = x**3
y3 = y1 + y2
# 反向传播计算梯度
y3.backward()
print("x 的梯度:", x.grad)
```
